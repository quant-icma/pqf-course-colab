## Topic

The goal of this project is to price a number of options using the Monte Carlo method.
We focus on plain (no early exercise), barrier and Asian options using the Black-Scholes
Bachelier, and Heston models. You should pay particular attention to your design. 
Make sure your project is flexible enough. For example, if later you want to 
introduce additional models or payoffs then your program should be extendable
to handle such future requirements.

You should implement the following payoffs:

- European call and put
- Up-and-Out option
- Down-and-Out option
- Up-and-In option
- Down-and-In option
- Asian call and put
- Basket call and put

with the following models:

- Bachelier
- Black-Scholes
- Heston

Compare at least two different algorithmic designs. For example, there are 3 common approaches:
- loop over the time steps and paths (double loop)
- loop over the time steps and process all paths together in a vectorized approach at every time step
- loop over the paths and process a full path in each iteration
