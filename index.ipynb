## Overview

This course provides an introduction to programming with the Python programming language using modern development environments.
The focus is on using Python in a financial engineering context, as a tool to implement and explore derivatives models. 
We do not assume any previous programming experience, but do assume some finance and derivatives modelling knowledge. 
The course is designed as a complement to the 'Derivatives Modelling' and 'Advanced Derivatives Modelling' courses at the ICMA centre.
Many of the hands-on consists in implementing models and conducting numerical experiments with models covered in those courses.

## Learning Objectives

By taking this course, you expect to:

  - become with programming with Python
  - learn software engineering skills required to develop small-scale real-world projects
  - get your 'hands dirty' at coding financial models
  - gain some experience at testing and debugging code

## Prerequisites

Intermediate knowledge of finance and derivatives in particular, 
at the level of an MSc in Quantitative Finance or Financial Engineering.
The course is designed to complement the 'Derivatives Modelling' 
and 'Advanced Derivatives Modelling' courses at the ICMA centre. 
This is approximately equivalent to the level of Joshi's book,
'The Concepts and Practice of Mathematical Finance' or Shreve's two-volume
'Stochastic Calculus for Finance'.

## Course Format

The course consists of:

  - 8 live introduction sessions
  - 10 pre-recorded lectures of 1 hour and 30 minutes
  - 6 graded individual programming assignments  
  - 1 final group project
